```python
import pandas as pd
import altair as alt

# Read the data from the Excel file into a DataFrame
df = pd.read_excel('SFTT.xlsx')

df = df.rename(columns={'Common Name': "CommonName"})
# Calculate the tree species counts
species_counts = df['CommonName'].value_counts().reset_index()
species_counts.columns = ['CommonName', 'Count']

# Select the top 10 tree species
top_10_species = species_counts.head(10)
selection = alt.selection_interval()

# Create the bar chart for top 10 tree species
bar_chart = alt.Chart(top_species).mark_bar().encode(
    x=alt.X('Count:Q', title='Tree Count'),
    y=alt.Y('CommonName:O', title='Tree Species').sort('-x'),
    color=alt.condition(
        alt.datum.CommonName == 'White Poplar',  # If the country is "US" this test returns True,
        alt.value('red'),     # highlight a bar with red.
        alt.value('lightgrey')),
    tooltip=['CommonName', 'Count']
).add_params(selection
).properties(
    title='Top Tree Species in Boston'
)


# Display the chart
bar_chart 
```
